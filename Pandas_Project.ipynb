```python
#importing the library
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv('movieset.csv')
data.head(5)

#plotting the line chart
# Line chart is used to represent a relationship between two data X and Y
plt.plot(data['IMDB_Rating'])

#adding title to plot
plt.title("Line Chart")

#setting the X and Y Lables
plt.ylabel('\Rating')
plt.xlabel('Index')
plt.show()
```

```python
# Bar Plot
# A graph that represents the category of data with rectagular bars with lengths and heights
df = data.head(10)

# Extracting data from data frame
Genre = df['Genre']
Rating = df['IMDB_Rating']

# Figue size
fig, ax = plt.subplots(figsize = (16,9))

# Set the axis range
ax.set(ylim=(7,9.5))

# Add plot title
ax.set_title('Movies type and their Ratings', loc = 'center', )

# Plotting the bar graph
plt.bar(Genre, Rating, color = 'Cyan',width = 0.4)
```

```python
# Histogram
df = data.head(50)
Meta = df['Meta_score']
num_bins=5
plt.hist(Meta, num_bins, facecolor='blue', alpha=0.5)

#Labels
plt.xlabel('Meta_Score')
plt.ylabel('Number of Movies')
plt.title('Meta Score Histogram')

#Range
plt.ylim(0, 10)
plt.xlim(50, 110)
```

```python
# Scatter plot
# Used to observe relationship between two variables and use dots
df = data.head(100)

# extracting data from data frame
Meta = df['Meta_score']
Rating = df['IMDB_Rating']

plt.scatter(Rating, Meta, c = 'blue')
```
